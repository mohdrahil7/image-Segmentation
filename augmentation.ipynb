```python
from keras.datasets import mnist
from matplotlib import pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
for i in range(0,9):
    plt.subplot(330+1+i)
    plt.imshow(x_train[i], cmap=plt.get_cmap("gray"))
plt.show()
```

```python
from keras.preprocessing.image import ImageDataGenerator
x_train=x_train.reshape(x_train.shape[0],28,28,1)
```

```python
datagen=ImageDataGenerator(rotation_range=90)
datagen.fit(x_train)
for x in datagen.flow(x_train,batch_size=10):
    for i in range(0,9):
        plt.subplot(330+1+i)
        plt.imshow(x[i].reshape(28,28))
    plt.show()
    break
```

```python
datagen=ImageDataGenerator(width_shift_range=0.2,height_shift_range=0.2)
datagen.fit(x_train)
for x in datagen.flow(x_train,batch_size=9):
    for i in range(0,9):
        plt.subplot(330+1+i)
        plt.imshow(x[i].reshape(28,28))
    plt.show()
    break
    
```

```python
datagen=ImageDataGenerator(horizontal_flip=True)
datagen.fit(x_train)
for x in datagen.flow(x_train,batch_size=9):
    for i in range(0,9):
        plt.subplot(330+i+1)
        plt.imshow(x[i].reshape(28,28))
    plt.show()
    break
```

```python
datagen=ImageDataGenerator(vertical_flip=True)
datagen.fit(x_train)
for x in datagen.flow(x_train,batch_size=9):
    for i in range(0,9):
        plt.subplot(330+i+1)
        plt.imshow(x[i].reshape(28,28))
    plt.show()
    break
```

```python
#shifting the image by a parameter randomly in left or right direction
datagen=ImageDataGenerator(width_shift_range=0.2)
datagen.fit(x_train)
for x in datagen.flow(x_train,batch_size=9):
    for i in range(0,9):
        plt.subplot(330+i+1)
        plt.imshow(x[i].reshape(28,28))
    plt.show()
    break
```

```python
#shifting the image by a parameter randomly in up or down direction
datagen=ImageDataGenerator(height_shift_range=0.2)
datagen.fit(x_train)
for x in datagen.flow(x_train,batch_size=9):
    for i in range(0,9):
        plt.subplot(330+i+1)
        plt.imshow(x[i].reshape(28,28))
    plt.show()
    break
```

```python
#zca whitening 
datagen = ImageDataGenerator(zca_whitening=True)
datagen.fit(x_train)
for x in datagen.flow(x_train, batch_size=9):
	for i in range(0, 9):
		plt.subplot(330 + 1 + i)
		plt.imshow(x[i].reshape(28, 28))
	plt.show()
	break
```
