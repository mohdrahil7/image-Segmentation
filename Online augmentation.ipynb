```python
from keras.datasets import mnist
from matplotlib import pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(x_train.shape[0],28,28,1)
x_test=x_test.reshape(x_test.shape[0],28,28,1)
```

```python
x_train.shape
```

```python
x_train.dtype
```

```python
#image before normalization
print(x_train[0])
plt.imshow(x_train[0].reshape(28,28))
```

```python
#data normalization
x_train=x_train.astype('float32')/255
x_test=x_test.astype('float32')/255
print(x_train[0])
```

```python
x_train.dtype
```

```python
x_test.dtype
```

```python
plt.imshow(x_train[0].reshape(28,28)) #image after normalization
```

```python
#before changing the data to the catogirical
print(y_train)
```

```python
print(y_train[0])
```

```python
#after changing to the categorical data
from keras.utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
print(y_train)
```

```python
print(y_train[0])
```

```python
#training the basic model on mnist data set 
from keras import models,layers
```

```python
model =models.Sequential()
model.add(layers.Conv2D(filters=6,kernel_size=(5,5), input_shape=(28,28,1) ,padding='valid', activation='relu',strides=(1,1) ))
model.add(layers.AveragePooling2D(pool_size=(2,2),strides=(2,2)))
model.add(layers.Conv2D(filters=16, kernel_size=(5,5), strides=(1,1),padding='valid',activation='relu' ))
model.add(layers.AveragePooling2D(pool_size=(2,2),strides=(2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(120,activation='relu'))
model.add(layers.Dense(84,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#model without using the data augmentation techniques
model.fit(x_train,y_train,batch_size=30,epochs=20,validation_split=(0.2),verbose=1)
```

```python
#using the online data augmentation using ImageDatagenerator
from keras.preprocessing.image import ImageDataGenerator
datagen=ImageDataGenerator(rotation_range=90,width_shift_range=0.2,height_shift_range=0.2,horizontal_flip=True,
                           vertical_flip=True,featurewise_center=False,samplewise_center=False,zca_whitening=False)
datagen.fit(x_train)
```

```python
model.summary()
```

```python
model.fit_generator(datagen.flow(x_train,y_train,batch_size=256),epochs=20,verbose=1,steps_per_epoch=(len(x_train)/256))
```
